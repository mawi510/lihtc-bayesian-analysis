## 2000-2026 CPI Less Housing
Source: https://fred.stlouisfed.org/series/CUUR0000SA0L2

```python
inflation = pd.read_csv('../data/raw/CUUR0000SA0L2.csv')
```

```python
inflation.head()
```

```python
#Lets rename the columns to make this easier to use
```

```python
inflation.columns = ['observation_date', 'cpi_index']
```

```python
#Lets also convert the observation to a datetime
```

```python
inflation['observation_date'] = pd.to_datetime(inflation['observation_date'])
```

```python
#Now let's get an average of monthly CPI measurements
#First let's get year
inflation['year'] = inflation['observation_date'].dt.year
```

```python
#Now we can annualize our CPI measurements
inflation_annual = inflation.groupby(['year'])['cpi_index'].mean().reset_index()
```

```python
inflation_annual
```

```python
# Save
inflation_annual.to_csv('../data/processed/inflation_annual_2000_2026.csv', index=False)
print('Saved!')
```
